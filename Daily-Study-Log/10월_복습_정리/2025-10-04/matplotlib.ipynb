```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 마커 
# "b" = 선
# "ro" = 빨간색 원
# marker="s" = 선, 모양
# marker="D" = 선, 마름모
# marker="$A$" = 선, 원하는 마커
plt.plot([4,5,6], "b")
plt.plot([3,4,5], "ro")
plt.plot([2,3,4], marker="s")
plt.plot([1,2,3], marker="D")
plt.plot([0,1,2], marker="$A$")
plt.show()
```

```python
# loc = title 위치
# pad = title과 graph 간격 

plt.plot([1,2,3,4],[1,4,6,7])
plt.xlabel('x_label')
plt.ylabel('y_label')
plt.title('graph title', loc='center', pad=10)
```

```python
# matplotlib 막대 그래프 bar
# xticks : 눈금
x = [1,2,3]
year = [2022,2023,2024]
value = [100, 500, 1000]

plt.bar(x, value, color= ['r','g','b'], width =0.4)
plt.xticks(x, year)

plt.show()


```

```python
# matplotlib 산점도 그래프 scatter
# numpy 와 random 모듈을 활용해서 데이터 생성

# 시드 고정 계속 같은 결과가 나오도록
np.random.seed(42)

# 0 ~ 1사이의 균등한 난수를 생성
n = 30
x = np.random.rand(n)
y = np.random.rand(n)

# scatter의 파라미터 값으로 넘겨줄 sixe와 color 정의
# size는 산점도의 크기
# color는 색상
size = (np.random.rand(n) * 20)**2
color = np.random.rand(n)

# x,y value와 s = size , c = color 
plt.scatter(x, y, s=size, c=color)
plt.show()
```

```python
# matplotlib subplot 그래프 
# 같은 도화지에 다른 그래프를 그리고 싶을 때 사용
x = np.random.rand(30)
y = np.random.rand(30)

# subplot
# nrows=2, ncols=1, index=1
# 같은 도화지에 행이 2개가 있고, 열은 1개가 있는데 그 중에서 첫 번째
plt.subplot(2, 1, 1)
plt.scatter(x,y)
plt.title('1st graph')

# subplot 
# 두 번째 그래프
plt.subplot(2, 1, 2)
plt.plot(x,y, 'o-')
plt.title('2st graph')

# tight_layout() 함수는 모서리와 서브플롯의 모서리 사이의 여백을 설정
plt.tight_layout()
plt.show()
```

```python
# subplots 한 도화지 안에 x축을 공유
# subplots를 호출하면 figure, graph 객체 생성

x = [2022,2023,2024]
y1 = [1,4,7]
y2 = [3,7,20]

# subplots(nrows, ncols, figsize=(가로크기, 세로크기))
fiq, ax1 = plt.subplots()

# -s(solid line style), alpha(투명도)
ax1.plot(x, y1 ,'-s', color = 'green', markersize=7, linewidth = 5, alpha=0.7)

# 같은 x축을 공유하지만 새로운 y축을 생성
ax2 = ax1.twinx()
ax2.bar(x, y2, color='deeppink', alpha=0.7, width=0.7)

plt.show()
```

```python
# 넷플릭스 데이터 불러오기
netflix = pd.read_csv(f"netflix_titles.csv", encoding= 'utf-8')
```

```python
# netflix 색상 팔레트
sns.palplot(['#221f1f','#b20710','#e50914','#f5f5f1'])

# loc = title 위치
# fontfamily = 폰트
# fontsize = 폰트 크기
# y = title 위치 조정
plt.title("Netflix brand palette", loc= 'left', fontfamily='serif', fontsize=15, y=1)
plt.show()
```

```python
netflix['type'].value_counts()
```

```python
# 전치 변환
ratio = pd.DataFrame(netflix['type'].value_counts()).T
ratio
```

```python
# pie plot 통해 비율 시각화
plt.figure(figsize=(5,5))

plt.pie(ratio, labels= ratio.columns, autopct='%.0f%%', startangle=100,
        explode=[0.05, 0.05], shadow=True, colors= ['#b20710', '#221f1f'])

plt.suptitle("Movie & TV distribution", fontfamily='serif', fontsize=15, fontweight='bold')
plt.title("We see more movies than TV shows on Netflix.", fontfamily= 'serif', fontsize=12)

plt.show()
```
