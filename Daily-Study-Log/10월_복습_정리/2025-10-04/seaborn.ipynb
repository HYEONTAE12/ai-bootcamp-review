```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# seaborn 공식사이트
# http://seaborn.pydata.org/
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
# sns regplot 그래프
# x축 value , y축 value, data =  , ax = 몇 번째 그래프, fix_reg= True 선형 회귀선 표시

fig = plt.figure(figsize=[15,5])
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)



# 산점도 선형 회귀선 표시 fit_reg = True(default)
sns.regplot(x='age', y='fare', data=titanic, ax=ax1)

# 산점도 선형 회귀선 미표시 fit_reg = False
sns.regplot(x='age', y='fare', data=titanic, ax=ax2, fit_reg=False)

plt.tight_layout()


plt.show()
```

```python
# sns displot 히스토그램 그래프와 커널 밀도 그래프 동시에
# kde = False(커널 밀도 표시 x)
# kde = True(커널 밀도 표시 O)
# hist = False(히스토그램 표시 x)

fig = plt.figure(figsize=[15,5])
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

# 히스토그램과 커널 밀도 전부표시
sns.histplot(titanic['fare'],kde= True, ax=ax1)

# 커널 밀도 그래프 미표시
sns.histplot(titanic['fare'],kde = False ,ax=ax2)

# 히스토그램 그래프 미표시
sns.kdeplot(titanic['fare'] ,ax=ax3)

# 제목 설정
ax1.set_title('titanic fare - hist + kde')
ax2.set_title('titanic fare - histplot')
ax3.set_title('titanic fare - kdeplot')

plt.tight_layout()
plt.show()
```

```python
# sns stripplot 
# 이산형 변수의 분포 - 데이터 분산 미고려
# x축 변수, y축 변수, 데이터 셋, axe 객체

# sns swarmplot 
# 이산형 변수의 분포 - 데이터 분산 고려
# x축 변수, y축 변수, 데이터 셋, axe 객체, hue= 색상 구분

# figsize 설정
fig = plt.figure(figsize=[15,5])
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)


sns.stripplot(x='class', y= 'age', data= titanic ,ax=ax1)
sns.swarmplot(x='class', y= 'age', data= titanic ,ax=ax2, hue ='sex')

# 제목 지정
ax1.set_title('Stripplot')
ax2.set_title('Swarmplot')

# 그래프 시각화
plt.show()
```

```python
# sns countplot 
# 각 범주의 속하는 데이터의 개수를 막대 그래프로 시각화

fig = plt.figure(figsize=[15,5])
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

# x축 변수, 데이터 셋, axe 객체, 1번째 그래프
sns.countplot(x='class', palette='Set1', data=titanic, ax=ax1)

# hue = 'who' 추가
sns.countplot(x='class', hue= 'who' , palette='Set2', data=titanic, ax=ax2)

# dodge= False 하나의 막대그래프에 표시
sns.countplot(x='class', hue= 'who' , palette='Set3', dodge=False, data=titanic, ax=ax3)


ax1.set_title('titanic class')
ax2.set_title('titanic class - who')
ax3.set_title('titanic class - who(stacked)')

plt.show()
```

```python
# sns pairplot 
# 여래개의 컬럼의 관계를 시각화

titanic_pair = titanic[['age', 'pclass', 'fare']]

sns.pairplot(titanic_pair)
```
