```python
# koEDA 데이터 증강 라이브러리 한국어 버전
from koeda import EasyDataAugmentation

def augment_text_data_with_EDA(text,repetition):
    eda = EasyDataAugmentation(
        morpheme_analyzer="Okt"
    )

    result = eda(text, p=(0.5,0.5,0.5,0.5), repetition=repetition)

    print("원문 :", text)
    print("--"*100)
    for i in range(repetition):
        print(f"증가문 {i+1}", result[i])
        
```

```python
augment_text_data_with_EDA(ex_data, 3)
```

```python
# AEDA 사용해보기 
from koeda import AEasierDataAugmentation

def augment_text_data_with_AEDA(text, repetition):
    aeda = AEasierDataAugmentation(morpheme_analyzer="Okt")
    result = aeda(text, p=0.3, repetition=repetition)

    print("원문 : ", text)
    print("-"*100)

    for i in range(repetition):
        print(f" 증가문{i+1}", result[i])
```

```python
augment_text_data_with_AEDA(ex_data, 2)
```

```python
# Back translation 사용해보기
# 한국어 -> 영어 -> 한국어
# 기존 텍스트를 영어로 번역 하고 다시 한국어로 번역함으로써 문장은 다르지만 원문과 비슷한 자연스러운 문장 생성이 가능함.
from transformers import pipeline

# 한국어 → 영어 -> 한국어
# 멀티모달 방식
translator = pipeline("translation", model="facebook/m2m100_418M")

def back_translation(text: str) -> str:
    # 1단계: 한국어 → 영어
    en = translator(text, src_lang="ko", tgt_lang="en")[0]['translation_text']
    # 2단계: 영어 → 한국어
    ko = translator(en,   src_lang="en", tgt_lang="ko")[0]['translation_text']
    return ko
```

```python
print(back_translation(ex_data))
```
