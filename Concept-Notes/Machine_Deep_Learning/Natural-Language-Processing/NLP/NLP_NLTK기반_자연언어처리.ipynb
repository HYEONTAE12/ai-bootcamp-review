```python
import nltk

nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
# pos_tag 함수를 통해 품사를 태깅할 수 있다.

sentence = """I can't believe it's already 2025! She said, "This is a state-of-the-art technology." 
The well-known scientist doesn't agree with the theory. However, they've made significant progress. 
Dr. Smith earned $50,000 last year. The temperature was -5.5°C yesterday."""


token = nltk.word_tokenize(sentence)
tagged = nltk.pos_tag(token)

for i, j in tagged:
    print(f"{i} - {j}")
```

```python
# 원하는 품사 word만 추출하고 싶을 때
allnoun = []
for word, pos in tagged:
    if pos in ['NN', 'NNP']:
        allnoun.append(word)


print("형태소 분석 결과: ", tagged)
print("명사만 추출 : ", allnoun)

allverb = []
for word, pos in tagged:
    if pos in ['VBZ', 'VBD']:
        allverb.append(word)

print('동사만 추출 : ', allverb)
```

```python
# 구문 분석
import nltk
import konlpy

sentence = '만 6세 이하의 초등학교 취학 전 자녀를 양육하기 위해서는'
words= konlpy.tag.Twitter().pos(sentence)
words
```

```python
# RegexpParser : 정규표현식을 기반으로 청킹을 수행하는 클래스
# 청킹(chunking) : 품사 태그가 붙은 단어들을 의미 있는 덩어리로 묶는 잡업
grammar = """
NP: {<N.*>*<Suffix>?}
VP: {<V.*>*}
AP: {<A.*>*}
"""

parser = nltk.RegexpParser(grammar)
chunks = parser.parse(words)
print("# Print whole tree")
chunks.pprint()
```

```python
# 그래픽으로 표현
chunks.draw()
```

```python
# 전체 문장이 아닌 부분 트리를 의미 
# 트리 안의 작은 트리들

# 사용하는 이유: 특정 태깅만 뽑고 싶을 때

for subtree in chunks.subtrees():
    if subtree.label() == "NP":
        print(' '.join((e[0] for e in list(subtree))))
        subtree.pprint()
```

```python
import pandas as pd
from tqdm import tqdm
from collections import Counter

import matplotlib.pyplot as plt
from wordcloud import WordCloud

import nltk
from nltk.corpus import stopwords
from nltk.tag import pos_tag
from nltk import word_tokenize

nltk.download('book', quiet= True)
nltk.download('punkt_tab', quiet= True)
nltk.download('stopwords')


nltk.corpus.gutenberg.fileids()
```

```python
data = nltk.corpus.gutenberg.raw('austen-emma.txt').lower()
```

```python
words = word_tokenize(data)
```

```python
def extract_by_tag(word_list, tag_name):
    pos_list = pos_tag(word_list, tagset='universal')
    extract_list = []

    for word_tag in tqdm(pos_list):
        if tag_name in word_tag[1]:
            extract_list.append(word_tag[0])
    return extract_list

def sorting(sentence_list):
    count = Counter(sentence_list)
    words = (dict(count.most_common()))
    return words       
       
```

```python
nouns =extract_by_tag(words, 'NOUN')
print(f"명사 예시: {nouns[5:10]}")

adjs = extract_by_tag(words, 'ADJ')
print(f"형용사 예시 : {adjs[:5]}")
 
```

```python
stop = stopwords.words('english')
stop.extend(['mr.', 'mrs.'])

nouns = [token for token in nouns if not token in stop]
adjs = [token for token in adjs if not token in stop]

nouns_count = sorting(nouns)
adjs_count = sorting(adjs)
```

```python
def view_wordcloud(counter_class):
    wordcloud = WordCloud(background_color= 'white',
                          colormap = "Accent_r", width=2500, height=1500,
                          stopwords=stopwords
                            ).generate_from_frequencies(counter_class)
    
    plt.figure(figsize=(10,8))
    plt.imshow(wordcloud)
    plt.tight_layout(pad=0)
    plt.axis('off')
    plt.show()


    plt.figure(figsize=(10,5))
    plt.xlabel('keywords')
    plt.ylabel('counts')
    plt.xticks(rotation=70)
    for key, value in zip(list(counter_class.keys())[:20], list(counter_class.values())[:20]):
        plt.bar(key, value, color='lightgrey')
    plt.show()


    counter_df = pd.DataFrame({'word':counter_class.keys(), 'counter':counter_class.values()})
    display(counter_df)
    return counter_df
```

```python
nouns_df = view_wordcloud(nouns_count)
```

```python
adjs_df = view_wordcloud(adjs_count)
```
