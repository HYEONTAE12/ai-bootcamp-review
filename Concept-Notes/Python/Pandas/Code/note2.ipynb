```python
train_x, test_x, train_y, test_y = train_test_split(main_origin, target, random_state=42, test_size = 0.25)
```

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()
X_train = std.fit_transform(train_x)
X_test = std.transform(test_x)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

models = {
    'LogReg' : LogisticRegression(max_iter=1000),
    'RF' : RandomForestClassifier(n_estimators=200, random_state=42),
    'XGB': XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}
```

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
```

```python
results =[]
for name, model in models.items():
    model.fit(X_train, train_y)
    pred = model.predict(X_test)
    proba = model.predict_proba(X_test)[:,1]

    results.append({
        'model' : name,
        'acc': accuracy_score(test_y, pred),
        'f1' : f1_score(test_y, pred),
        'auc' : roc_auc_score(test_y, proba)
    })

df_results = pd.DataFrame(results).sort_values("auc", ascending=False)
df_results
```
